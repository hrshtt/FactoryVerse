```python
import json
from factorio_rcon import RCONClient

rcon_client = RCONClient("localhost", 27100, "factorio")
 
```

```python
# -- List all available tests
res = rcon_client.send_command('/c rcon.print(remote.call("test_runner", "list_tests"))')
print(res)
```

```python
# -- Run all tests
res = rcon_client.send_command('/c rcon.print(remote.call("test_runner", "run_all_tests"))')
print(res)
```

```python
for item in json.loads(res)['results']:
    if 'error' in item:
        print("#"*50)
        print(item['test_name'])
        print("="*50)
        print(item['error'])
```

```python
# -- Run a specific category
res = rcon_client.send_command('/c rcon.print(remote.call("test_runner", "run_category", "agent"))')
print(res)
```

```python
# -- Run a single test
res = rcon_client.send_command('/c rcon.print(remote.call("test_runner", "run_test", "agent.test_create"))')
print(res)
```

```python
# -- Check async test status (if any are pending)
res = rcon_client.send_command('/c rcon.print(remote.call("test_runner", "get_pending_tests"))')
print(res)
```
